2.) Supervised Learning: You have fully labeled data of your training samples meaning you have access to the output classes of your train instances. So your model is trained to learn the weights of your features that will minimize the loss between the model's estimated output and the correct output for every training sample. An example would be classification of cars where each training sample is an image of the car and the output of it is the label of the car. 

Unsupervised Learning: In this type of learning, the model doesn't know the correct output of the training instances. So the model has to find patterns in the data without knowing the labels of the data. It can learn the organization of the data by clustering for example the training samples into clusters where the samples in a cluster would be more similar to each other than to any other samples in other clusters according to a metric. Another example of an unsupervised learning model would be an autoencoder which takes input data and compresses it then tries to create the input data again from the compressed version of it. An example of clustering would be clustering set of individuals according to some given features. 

Reinforcement Learning: In this type of learning, the model has to make a decision at every step where the previous decisions influence the decision at that step. The decision at the current step also  influences the next input you will get. In reinforcement learning, the model is learning which behaviors to reward and which behaviors to punish and it does that dynamically. An example would be teaching a robot how to walk in a coordinate space.

This dataset consists of features for every training sample along with the correct class output of the sample. The model uses these trained samples to fit its weights to minimize the loss between the model's estimated output and the correct output for the training samples. Test dataset consist of labeled test samples with the correct output but however aren't used to train the model. These test samples are given to the model as input to the model learned from the training samples. The model's estimated output on the test samples is compared to the correct class outputs of the test datasets to see how well it can generalize to unseen data and measure the accuracy of generalization according to some metric.
